```python
import duckdb
import pandas as pd
import os

# Connect to a database (in-memory by default)
conn = duckdb.connect("hcr_comparison.db")

# Define export to Excel sheet functionality
def save_to_excel(query: str, filename: str):
    results = conn.execute(query).fetchdf()
    output_dir = '/Users/work/Documents/prototype-analytics-platform/src/silver/POC'
    
    # Combine directory and filename properly
    full_path = os.path.join(output_dir, filename)
    
    print(f"Saving results to: {full_path}")
    results.to_excel(full_path, index=False, engine='openpyxl', na_rep='NULL')
    print(f"Successfully saved {len(results)} rows to {filename}")

# Query to display results from table
query = """
                SELECT *
                FROM 
                    agricultural_sciences_feb_highly_cited_only
                """

query_results = conn.execute(query).fetchdf()

# Display audit results from query
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)        # Auto-detect display width
pd.set_option('display.max_rows', 20)       # Show first/last 10 rows (adjust as needed)
pd.set_option('display.max_colwidth', 100)  # Limit column width for readability

display(query_results)

# Save to XLSX
# Input excel spreadsheet name below
save_to_excel(query, 'test_scopus_query_results_v1.xlsx')

# Close the connection when done
conn.close()
```
